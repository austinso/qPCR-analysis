```python
exptID = "/FD0040"

exptFolder = "/Users/austinso/Box Sync/Experimental_Plans/FD (FFPE DNA)" + exptID
dataFolder = "/Users/austinso/Box Sync/Pilot Runs Raw Data/Kapa_HG" + exptID

z_cutoff = 3
```

Data from the Q5 exported table (analyzed) is merged to the plate map that was used for the experiment to allow unstructured slurping of the data (both as excel files).

Q5 qPCR data is exported as a single file with multiple worksheets in to the "Pilot Runs Raw Data/Kapa_HG" folder within the experimental ID subfolder. The relevant worksheet is the "Results" worksheet (generated by the Q5 software), and taking the "Well Position" and the "CT" column values.

```python
files = glob.glob(dataFolder+"/*.xls")
list_of_dfs = [pd.read_excel(f,sheet_name="Results", header=None, usecols="B,I", dtype={'I':np.float64}, na_values='CT').dropna(axis=0, subset=[1]) for f in files]

p = re.compile(r'(\w+).xls$')
for dataframe, f in zip(list_of_dfs, files):
  dataframe[2] = re.search(p,f).group(1)

qPCRdata = pd.concat(list_of_dfs, ignore_index=True)
qPCRdata = qPCRdata.rename(columns={0:'Well Position',1:'CT',2:'plateID'}).replace('Undetermined', np.nan)

qPCRdata.head()
```

"platemap" is a file within the "Experimental Plans/FD (FFPE DNA)" folder within the experimental ID subfolder that outlines the actual sample allocation across the plate. Additional data files are utilized within this folder to parse information and provide basis for analyses.

The sample IDs within the platemap should have the following general structure with each field separated by a "-" in order to parse properly:

lysate/eluate_ID-assay-dilutionFactor-replicate

Note: replicate refers to any technical replicates rather than measurement replicates. 

Standards are currently identified by:
stdn, where n = {1-5}
Note: it may be better to unify this by having the same structure as the sampleIDs

```python
xlsx = pd.ExcelFile(exptFolder+"/platemap.xlsx")
p = re.compile(r'platemap')
list_of_dfs = [ xlsx.parse(sheet) for sheet in xlsx.sheet_names if re.search(p,sheet)]
sheets = [sheet for sheet in xlsx.sheet_names if re.search(p,sheet)]

platemaps = pd.DataFrame()
p = re.compile(r'_(\w+)$')
for f,sampleID in zip(sheets,list_of_dfs):
    sampleID = sampleID.stack().reset_index()
    sampleID['Well Position']=sampleID["level_0"].map(str) + sampleID["level_1"].map(str)
    sampleID.drop(['level_0', 'level_1'], axis=1)
    sampleID.rename(columns={0: 'sampleName'}, inplace=True)

    name = sampleID.sampleName.str.split(pat="-",expand=True)
    sampleID['name']=name[0]
    sampleID['assay']=name[1]
    sampleID['factor']=name[2].astype('float')
    sampleID['detail']=name[3]

    sampleID['plateID'] = re.search(p,f).group(1)
    
    platemaps=pd.concat([platemaps,sampleID[['plateID','Well Position','name','assay','factor','detail']]],ignore_index=True).fillna(1)

platemaps.head()
```

Now I can join the two tables to create the master data table that draws from the experimental plans, and ITP runs.

```python
dataTable = pd.merge(platemaps,qPCRdata,how='left',on=['plateID','Well Position'])
dataTable.head()
```

========================================

Processing standards for analysis

Note: 
1	2500 pg/uL	770 copies/uL
2	625 pg/uL	189.5 copies/uL
3	156 pg/uL	47.25 copies/uL
4	39.1 pg/uL	11.75 copies/uL
5	9.77 pg/uL	3 copies/uL

```python
standardData = dataTable[dataTable.name.str.contains(r'std', regex=True, case=False)]
standardData['logConc']=np.log10(standardData['factor'])
standardData.head()
```

```python
standards = standardData.groupby(['plateID','assay','name'])['CT'].describe().reset_index()
standards.head()
```

```python
standardData = pd.merge(standardData, standards[['plateID','assay','name','mean','std','count']], how='left',on=['plateID','assay','name'])
standardData.head()
```

Introduce outlier test to remove standard replicates that are outside of distribution. This is to remove any egregious outliers based on a z-score > 3 (99.9% CI) that suggest a technical error.

```python
standardData['z_mean']=np.abs((standardData['CT'] - standardData['mean'])/(standardData['std']/np.sqrt(standardData['count'])))
standardData['z_mean']=np.where(standardData['z_mean']>z_cutoff,np.nan,standardData['z_mean'])
standardData = standardData.drop(columns=['mean','std','count'])
standardData.head()

```

```python
standards = standardData.dropna(subset=['z_mean']).groupby(['plateID','assay','name','logConc'])['CT'].describe().reset_index()
standards
```

```python
ax = plt.gca()
plt.legend()
for plate, group in standardData.dropna(subset=['z_mean']).groupby('plateID'):
    group.plot(x='logConc', y='CT', kind='scatter', ax=ax, label="plate_"+plate)
```

```python
standardsFit=pd.DataFrame()
fit=pd.DataFrame()

ax = plt.gca()

for plate, group in standardData.dropna(subset=['z_mean']).groupby('plateID'):

    X = group['logConc'].values.reshape(-1,1)
    y = group['CT'].values.reshape(-1,1)
    
    lm = linear_model.LinearRegression()
    model = lm.fit(X,y)

    group['predicted']=lm.predict(X)
    
    fit['intercept'] = lm.intercept_
    fit['slope'] = lm.coef_
    fit['R^2'] = lm.score(X, y)
    fit['plateID'] = plate
        
    standardsFit = standardsFit.append(fit,ignore_index=True)
    
    group.groupby(['plateID','assay','logConc'])['CT'].describe().reset_index().plot.scatter(x='logConc', y='mean', yerr='std', marker='o',label="plate_"+plate, ax=ax)
    group.groupby(['plateID','assay','logConc'])['predicted'].describe().reset_index().plot(x='logConc', y='mean', kind='line',  label="plate_"+plate+" fit", ax=ax)
    #figure.plot.scatter(x='logConc', y='mean', color=c[plate], marker='o',label="plate_"+plate, ax=ax)
    
```

Processing samples for analysis.

Select only data from the samples and convert to concentration using the plate specific standards

Introduce outlier test to remove replicates that are outside of distribution. This is to remove any egregious outliers based on a z-score > 3 (99.9% CI) that suggest a technical error:
1. filter out by dilution factor to allow assessment of putative inhibition relief
2. filter out by corrected concentration to obtain aggregate estimate of input quantities

```python
sampleData = dataTable[~dataTable.name.str.contains(r'(std)|(ntc)', regex=True, case=False)]
```

```python
samples_by_factor = sampleData.groupby(['plateID','assay','name','factor','detail'])['CT'].describe().reset_index()
samples_by_factor.head()
```

```python
sampleData = pd.merge(sampleData, samples_by_factor[['plateID','assay','name','factor','detail','mean','std','count']], how='left',on=['plateID','assay','name','factor','detail'])
sampleData.head()
```

```python
sampleData['z_factor']=np.abs((sampleData['CT'] - sampleData['mean'])/(sampleData['std']/np.sqrt(sampleData['count'])))
sampleData['z_factor'] = np.where(sampleData['z_factor']>z_cutoff,np.nan,sampleData['z_factor'])
sampleData = sampleData.drop(columns=['mean','std','count'])
sampleData.head()
```

```python
sampleData = pd.merge(sampleData,standardsFit,how='left',on=['plateID'])
```

```python
sampleData['conc']=10**((sampleData['CT']-sampleData['intercept'])/sampleData['slope'])
```

```python
sampleData['correctedConc']=sampleData['conc']*sampleData['factor']
sampleData = sampleData.drop(columns=['intercept','slope','R^2'])
sampleData.head()
```

Now filter data by corrected concentrations to allow for aggregate mean

```python
samples_by_name = sampleData.groupby(['plateID','assay','name','detail'])['correctedConc'].describe().reset_index()
samples_by_name.head()
```

```python
sampleData = pd.merge(sampleData, samples_by_name[['plateID','assay','name','detail','mean','std','count']], how='left',on=['plateID','assay','name','detail'])
sampleData.head()
```

```python
sampleData['z_name']=np.abs((sampleData['correctedConc'] - sampleData['mean'])/(sampleData['std']/np.sqrt(sampleData['count'])))
sampleData['z_name']=np.where(sampleData['z_name']>z_cutoff,np.nan,sampleData['z_name'])
sampleData = sampleData.drop(columns=['mean','std','count'])
sampleData.head()
```

```python
summary = sampleData.groupby(['plateID','name','detail','assay','factor'])['correctedConc'].describe()
summary.head()
```

```python
summary.plot(y='mean', kind='bar',yerr='std',figsize=(20,4))
```

Hmmm...the more I think about this, I think that this sampleData table should be pushed to the expSummaryFile, and then the analysis should be taken from there. However, the issue is in how to account for the granular-level data that allows you to go to a deep-dive in the data i.e. demonstrating inhibition through dilutions, or looking at outliers, etc...



```python
lysateData = pd.merge(sampleData, expSummaryFile[['Lysate_ID','Input_Volume_ul']], how='inner', left_on='name', right_on='Lysate_ID',)
eluateData = pd.merge(sampleData, expSummaryFile[['Lysate_ID','Eluate_ID','Eluate_Volume_ul']], how='inner', left_on='name', right_on='Eluate_ID')
summaryData = pd.merge(lysateData,eluateData,how='outer')
summaryData.head()
```

```python
summaryData = pd.merge(summaryData,lysateLabelFile[['ID','Block_ID','Note']], how='inner', left_on='Lysate_ID', right_on='ID')
summaryData.head()
```

```python
summaryData['mass'] = np.where(summaryData['Eluate_ID'].isnull(),(0.003*summaryData['correctedConc']*summaryData['Input_Volume_ul']), (0.003*summaryData['correctedConc']*summaryData['Eluate_Volume_ul']))
summaryData['type'] = np.where(summaryData['Eluate_ID'].isnull(),'lysate','eluate')
summaryData['Eluate_ID'] = np.where(summaryData['Eluate_ID'].isnull(),'none',summaryData['Eluate_ID'])
summaryData.head()
```

```python
summaryLysate = summaryData.dropna(subset=['z_name','z_factor']).groupby(['Lysate_ID','Eluate_ID'])['mass'].describe().sort_index().reset_index()
summaryLysate.head()
```

```python
summaryLysate['CV']=summaryLysate['std']/summaryLysate['mean']
summaryLysate['95CI']=summaryLysate['std']*1.96
```
